```python
import pandas as pd

agentes = pd.read_csv("agentes_valorant.csv")

agentes.head()
```

```python
agentes.columns
```

```python
agente_mais_usado = agentes.loc[agentes["Taxa de uso"].idxmax()]

agente_mais_usado
```

```python
import matplotlib.pyplot as plt

agentes.plot(
    x="Agente",
    y="Taxa de uso",
    kind="bar",
    title="Uso de agentes no Valorant",
    legend=False
)

plt.show()
```

```python
plt.savefig("images/agentes_mais_usados.png")
```
